### Друзья у людей с разными именами.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vk_names.csv', sep=';', header=None)
df.columns = ['uid', 'name', 'surname', 'friends']
df.dropna(inplace=True)
df = df[:500000]
```

```python
df.info()
```

```python
topnms = df.groupby('name').count().sort_values(['friends'], ascending=False).iloc[:20, 0]
plt.figure(figsize=(10, 6))
plt.grid(axis='x')
sns.barplot(x=topnms.values, y=topnms.index, orient='h')
plt.xlabel('Number of occurences')
plt.ylabel('Name')
plt.title('Most popular names (from {} people)'.format(df.shape[0]))
plt.savefig('topnames.png')
```

```python
df2 = df.drop(df[df['friends'] == 0].index)
#df2.drop(df2[df2['friends'] > np.percentile(df2['friends'], 95)].index, inplace=True)
df2.drop(df2[df2['friends'] > 2000].index, inplace=True)
df2 = df2.groupby('name', as_index=False).sum().loc[:, ['name', 'friends']]
df2 = df2.sort_values(by=['friends'], ascending=False).reset_index()
friends = pd.concat([df2.iloc[:19], df2[df2['name'] == 'Anton']])
friends.head()
```

```python
plt.figure(figsize=(10, 6))
plt.grid(axis='x')
sns.barplot(x=friends['friends'], y=friends['name'], orient='h')
plt.xlabel('Количество друзей')
plt.ylabel('Имя')
plt.title('Самый удачливые на друзей люди из выборки в 500000 человек. (сумма друзей на имя)')
plt.savefig('top_friends.png')
```
